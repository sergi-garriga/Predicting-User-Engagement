### Add polarity, topic and score columns to df

```python
df['topic'] = df_topic['topic']
df['topic_name'] = df_topic['topic_name']
```

```python
df['polarity'] = df_polarity['polarity']
df['polarity_score'] = df_polarity['polarity_score']
```

```python
df['score'] = df_score['score']
df['score_name'] = df_score['score_name']
df['rank'] = df_score['rank']
df['status'] = df_score['status']
```

# Clean data

```python
df['polarity'] = df['polarity'].str.replace(r'\D+', '', regex=True)
df['polarity'].value_counts()
```

```python
df['polarity'] = df['polarity'].astype('int')
```

```python
df.info()
```

### Number of relevant

```python
df['num_relevant'] = [len(df['token_lemma_relevant'][i].split()) for i in range(len(df))]
```
